```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Subcategory': ['X', 'X', 'Y', 'Y', 'X', 'X'],
        'Value': [10, 20, 30, 40, 50, 60]}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_table = df.pivot_table(index='Category', columns='Subcategory', values='Value', aggfunc='sum')

print(pivot_table)
```

```python
import matplotlib.pyplot as plt

# Plotting the pivot table data
pivot_table.plot(kind='bar', figsize=(8, 6))
plt.title('Pivot Table Visualization')
plt.ylabel('Sum of Values')
plt.xlabel('Category')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for plotting
data = [10, 20, 30, 40, 50]
categories = ['A', 'B', 'C', 'D', 'E']

plt.plot(categories, data, marker='o')

# Annotating a point
plt.annotate('Max value', xy=('E', 50), xytext=('D', 45),
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=12)

# Show plot
plt.show()
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# Using catplot
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
# hue
sns.boxplot(data=tips,y='day',x='total_bill',hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
sns.residplot(data=tips,x='total_bill',y='tip')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('marketing_sales.csv')

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['marketing_spend'], df['sales_revenue'])

# Annotating key points
for i, row in df.iterrows():
    if row['marketing_spend'] > 50000:
        plt.annotate(f"{row['marketing_spend']}", (row['marketing_spend'], row['sales_revenue']))

# Adding labels and title
plt.xlabel('Marketing Spend')
plt.ylabel('Sales Revenue')
plt.title('Marketing Spend vs Sales Revenue')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Creating the CSV file with sample data
data = {
    'marketing_spend': [5000, 35000, 50000, 15000, 45000, 70000, 10000, 60000, 55000, 30000],
    'sales_revenue': [220000, 280000, 110000, 250000, 320000, 95000, 300000, 290000, 180000, 210000]
}

df = pd.DataFrame(data)

# Save the data to a CSV file
df.to_csv('marketing_sales.csv', index=False)

# Step 2: Load the data (in case you want to reload it)
df = pd.read_csv('marketing_sales.csv')

# Step 3: Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['marketing_spend'], df['sales_revenue'], color='b', label='Data Points')

# Step 4: Annotate points where marketing spend exceeds $50,000
for i in range(len(df)):
    if df['marketing_spend'][i] > 50000:
        plt.annotate(f'${df["marketing_spend"][i]}', (df['marketing_spend'][i], df['sales_revenue'][i]),
                     textcoords="offset points", xytext=(0, 10), ha='center', fontsize=9, color='red')

# Step 5: Adding labels and title
plt.xlabel('Marketing Spend ($)', fontsize=12)
plt.ylabel('Sales Revenue ($)', fontsize=12)
plt.title('Marketing Spend vs Sales Revenue', fontsize=14)

# Step 6: Add grid
plt.grid(True)

# Step 7: Display the plot
plt.show()
```
