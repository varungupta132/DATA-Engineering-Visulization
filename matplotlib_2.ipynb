```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a simple dataset
data = {
    'SepalLengthCm': [5.1, 4.9, 4.7, 4.6, 5.0],
    'PetalLengthCm': [1.4, 1.4, 1.3, 1.5, 1.4]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Scatter plot of SepalLengthCm vs PetalLengthCm
# Using 'c' to map SepalLengthCm to colors
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c=df['SepalLengthCm'], cmap='jet',alpha = 0.1)

# Set labels for the axes
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Add color bar
plt.colorbar()
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.random.rand(10)
y4 = np.random.randn(1000)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plotting data in each subplot

# Top-left subplot: Line plot
axs[0, 0].plot(x, y1, label='Sine Wave', color='blue')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].legend()

# Top-right subplot: Scatter plot
axs[0, 1].scatter(x, y2, label='Cosine Wave', color='red')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].legend()

# Bottom-left subplot: Bar chart
axs[1, 0].bar(range(10), y3, color='green')
axs[1, 0].set_title('Random Bar Chart')

# Bottom-right subplot: Histogram
axs[1, 1].hist(y4, bins=30)
axs[1, 1].set_title('Random Normal Distribution')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a meshgrid
x = np.linspace(0.1, 5, 100)  # Avoid log(0) by starting from 0.1
y = np.linspace(0.1, 5, 100)
xx, yy = np.meshgrid(x, y)

# Define the function z
z = np.sin(xx) + np.log(xx)

# Create a figure and a 3D axis
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')

# Plot the surface
p = ax.plot_surface(xx, yy, z, cmap='jet')

# Add a colorbar
fig.colorbar(p, ax=ax, shrink=0.5, aspect=10)

# Set labels and title
ax.set_title("3D Surface Plot with Jet Colormap")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

# Show the plot
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='jet',c = xx)
fig.colorbar(p)
```
