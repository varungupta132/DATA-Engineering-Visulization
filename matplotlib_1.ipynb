### 2D Line plot

image.png

- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
        'Sales': [200, 250, 180, 300, 220]}
df = pd.DataFrame(data)

# Plot
plt.bar(df['Day'], df['Sales'])
plt.title("Average Sales by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [2, 4, 6, 8, 10]  # Y-axis values

# Plot
plt.plot(x, y)  # Line plot
plt.xlabel("X-axis Label")  # Label for X-axis
plt.ylabel("Y-axis Label")  # Label for Y-axis
plt.title("2D Graph Example")  # Title of the graph

# Show the plot
plt.grid(True)  # Adding a grid for clarity
plt.show()
plt.plot(x, y)
plt.yticks([2, 4, 6, 8, 10])  # Specify only the Y-axis ticks you want
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Tick Marks")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'batsman': ['Player1', 'Player2', 'Player3', 'Player4'],
    '2015': [10, 20, 30, 40],
    '2016': [15, 25, 35, 45],
    '2017': [20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Plotting with adjusted width and shifted positions
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='yellow', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='red', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='blue', label='2017')

# Setting x-axis labels
plt.xticks(np.arange(df.shape[0]), df['batsman'])

# Adjust xlim to ensure enough space for shifted bars
plt.xlim(-0.5, df.shape[0] - 0.5)

# Adding a title and labels
plt.title('Comparison of Batsmen for 2015, 2016, and 2017')
plt.xlabel('Batsman')
plt.ylabel('Score')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'batsman': ['Player1', 'Player2', 'Player3', 'Player4'],
    '2015': [10, 20, 30, 40],
    '2016': [15, 25, 35, 45],
    '2017': [20, 30, 40, 50]
}
df = pd.DataFrame(data)

# Plotting with adjusted width and shifted positions
plt.bar(np.arange(df.shape[0]) - 0.3, df['2015'], width=0.3, color='yellow', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.3, color='red', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.3, df['2017'], width=0.3, color='blue', label='2017')

# Setting x-axis labels
plt.xticks(np.arange(df.shape[0]), df['batsman'])

# Adjust xlim to ensure enough space for shifted bars
plt.xlim(-1, df.shape[0] +1)

# Adding a title and labels
plt.title('Comparison of Batsmen for 2015, 2016, and 2017')
plt.xlabel('Batsman')
plt.ylabel('Score')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (replace with your actual data)
df = {
    'batsman': ['Player1', 'Player2', 'Player3', 'Player4', 'Player5'],
    '2015': [10, 15, 20, 25, 30],
    '2016': [12, 18, 22, 28, 35],
    '2017': [14, 20, 25, 30, 40]
}
df = pd.DataFrame(df)

# Increased spacing between bars
spacing = 0.3  # Increase the spacing between bars
bar_width = 0.7  # Bar width

plt.bar(np.arange(df.shape[0]) - spacing, df['2015'], width=bar_width, color='yellow')
plt.bar(np.arange(df.shape[0]), df['2016'], width=bar_width, color='red')
plt.bar(np.arange(df.shape[0]) + spacing, df['2017'], width=bar_width, color='blue')

# Set x-ticks for batsman names
plt.xticks(np.arange(df.shape[0]), df['batsman'])

# Adjust x-axis limits to make space on both sides
plt.xlim(-1, df.shape[0] + 1)

# Show the plot
plt.show()

```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contribution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects,autopct='%0.1f%%',explode=[0,0.3,0,0,0])

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart

labels = ['Apple', 'Banana', 'Cherry', 'Date']  # Categories
sizes = [30, 40, 30,100]  # Proportions for each category
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']  # Colors for each slice

# Create a pie chart
plt.pie(sizes, labels=["a","b","c","d"], colors=colors, autopct='%0.1f%%', startangle=720)

# Add a title
plt.title('Fruit Distribution')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [15, 30, 45, 10]  # Pie chart data
labels = ['A', 'B', 'C', 'D']  # Labels for each slice
explode = (0.1, 0.001, 0.2, 0.03)  # "Exploding" the first slice (A)

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%')
plt.show()
```
